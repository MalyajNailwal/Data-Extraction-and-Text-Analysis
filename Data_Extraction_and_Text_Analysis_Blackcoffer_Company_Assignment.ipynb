Data extraction


```python
# Importing necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import os  # Import the 'os' module for creating folders
from google.colab import files
import nltk
import string
nltk.download('punkt')


# Specify the folder path
output_folder = '/content/drive/MyDrive/Data_Extraction_and_Text_Analysis_Blackcoffer_Company_Assignment/TitleText'

# Check if the folder already exists
if not os.path.exists(output_folder):
    # If it doesn't exist, create the folder
    os.makedirs(output_folder)
    print(f"Folder '{output_folder}' created successfully.")

url_id = 1
for i in range(0, len(df)):
    j = df.iloc[i].values

    # Set up headers to mimic a web browser
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}

    # Make a request to the URL
    page = requests.get(j[0], headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')

    # Extract content from the web page
    content = soup.findAll(attrs={'class': 'td-post-content'})

    # Check if content is not empty before accessing its index
    if content:
        content = content[0].text.replace('\xa0', "  ").replace('\n', "  ")
    else:
        content = ""  # Set content to an empty string if it's empty

    # Extract title from the web page
    title = soup.findAll(attrs={'class': 'entry-title'})

    # Check if title is not empty before accessing its index
    if title:
        title = title[16].text.replace('\n', "  ").replace('/', "")
    else:
        title = ""  # Set title to an empty string if it's empty

    # Combine title and content
    text = title + '.' + content
    text = np.array(text)
    text.reshape(1, -1)
    df1 = pd.Series(text)

    # Create the file path for saving the text file
    b = os.path.join(output_folder, f"{url_id}.txt")

    with open(b, 'a') as f:  # Creating and writing to the text file
        df1.to_csv(f, line_terminator=',', index=False, header=False)
        # Uncomment the line below to download the text file (commented out to avoid potential issues)
        # files.download(b)

    url_id += 1  # Increment the URL ID for the next iteration
```

```python
# Importing stop words files that are provided
StopWords_Auditor = pd.read_csv("/content/drive/MyDrive/Data_Extraction_and_Text_Analysis_Blackcoffer_Company_Assignment/StopWords/StopWords_Auditor.txt", header=None)
StopWords_Currencies = pd.read_csv("/content/drive/MyDrive/Data_Extraction_and_Text_Analysis_Blackcoffer_Company_Assignment/StopWords/StopWords_Currencies.txt", header=None, encoding="ISO-8859-1", sep='\t')
StopWords_DatesandNumbers = pd.read_csv("/content/drive/MyDrive/Data_Extraction_and_Text_Analysis_Blackcoffer_Company_Assignment/StopWords/StopWords_DatesandNumbers.txt", header=None)
StopWords_Generic = pd.read_csv("/content/drive/MyDrive/Data_Extraction_and_Text_Analysis_Blackcoffer_Company_Assignment/StopWords/StopWords_Generic.txt", header=None)
StopWords_GenericLong = pd.read_csv("/content/drive/MyDrive/Data_Extraction_and_Text_Analysis_Blackcoffer_Company_Assignment/StopWords/StopWords_GenericLong.txt", header=None)
StopWords_Geographic = pd.read_csv("/content/drive/MyDrive/Data_Extraction_and_Text_Analysis_Blackcoffer_Company_Assignment/StopWords/StopWords_Geographic.txt", header=None)
StopWords_Names = pd.read_csv("/content/drive/MyDrive/Data_Extraction_and_Text_Analysis_Blackcoffer_Company_Assignment/StopWords/StopWords_Names.txt", header=None)
```

```python
#creating func for removing stop words
def text_process(text):
    nopunc =[char for char in text if char not in punc or char not in [':',',','(',')','â€™','?']]
    nopunc=''.join(nopunc)
    txt=' '.join([word for word in nopunc.split() if word.lower() not in StopWords_Auditor])
    txt1=' '.join([word for word in txt.split() if word.lower() not in StopWords_Currencies])
    txt2=' '.join([word for word in txt1.split() if word.lower() not in StopWords_DatesandNumbers])
    txt3=' '.join([word for word in txt2.split() if word.lower() not in StopWords_Generic])
    txt4=' '.join([word for word in txt3.split() if word.lower() not in StopWords_GenericLong])
    txt5=' '.join([word for word in txt4.split() if word.lower() not in StopWords_Geographic])
    return ' '.join([word for word in txt5.split() if word.lower() not in StopWords_Names])
```

```python
#applying func for each row
c['abc']=c['abc'].apply(text_process)
```

```python
#negative list
length=negative.shape[0]
neg=[]
for i in range(0,length):
  nopunc =[char for char in negative.iloc[i] if char not in string.punctuation or char != '+']
  nopunc=''.join(nopunc)
  neg.append(nopunc)
```

```python
#positive list
length=positive.shape[0]
post=[]
for i in range(0,length):
   nopunc =[char for char in positive.iloc[i] if char not in string.punctuation or char != '+']
   nopunc=''.join(nopunc)

   post.append(nopunc)
```

```python
#importing tokenize library
from nltk.tokenize import word_tokenize
```

```python
txt_list=[]
length=c.shape[0]
for i in range(0,length):
  txt=' '.join([word for word in c.iloc[i]])
  txt_list.append(txt)
```

```python
#tokenization of text
tokenize_text=[]
for i in txt_list:

  tokenize_text+=(word_tokenize(i))
```

1.POSITIVE SCORE


```python
positive_score=0
for i in tokenize_text:
  if(i.lower() in post):
    positive_score+=1
print('postive score=', positive_score)
```

2.NEGATIVE SCORE


```python
negative_score=0
for i in tokenize_text:
  if(i.lower() in neg):
    negative_score+=1
print('negative score=', negative_score)
```

3.POLARITY SCORE

```python
#Polarity Score = (Positive Score â€“ Negative Score)/ ((Positive Score + Negative Score) + 0.000001)
Polarity_Score=(positive_score-negative_score)/((positive_score+negative_score)+0.000001)
print('polarity_score=', Polarity_Score)
```

4.SUBJECTIVITY SCORE

```python
#Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)
subjectiivity_score=(positive_score-negative_score)/((len(tokenize_text))+ 0.000001)
print('subjectivity_score',subjectiivity_score)
```

5.AVG SENTENCE LENGTH

```python
length=c.shape[0]
avg_length=[]
for i in range(0,length):
  avg_length.append(len(c['abc'].iloc[i]))
avg_senetence_length=sum(avg_length)/len(avg_length)
print('avg sentence length=', avg_senetence_length)
```

6.PERCENTAGE OF COMPLEX WORDS

```python
vowels=['a','e','i','o','u']
import re
count=0
complex_Word_Count=0
for i in tokenize_text:
  x=re.compile('[es|ed]$')
  if x.match(i.lower()):
   count+=0
  else:
    for j in i:
      if(j.lower() in vowels ):
        count+=1
  if(count>2):
   complex_Word_Count+=1
  count=0
```

```python
Percentage_of_Complex_words=complex_Word_Count/len(tokenize_text)
print('percentag of complex words= ',Percentage_of_Complex_words)
```

7.FOG INDEX

```python
#Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)
Fog_Index = 0.4 * (avg_senetence_length + Percentage_of_Complex_words)
print('fog index= ',Fog_Index )
```

8.AVG NUMBER OF WORDS PER SENTENCE

9.COMPLEX WORD COUNT

```python
vowels = {'a', 'e', 'i', 'o', 'u'}
count = 0
complex_Word_Count = 0
for i in tokenize_text:
    x = re.compile('es$|ed$', re.IGNORECASE)
    if x.match(i):
        count += 0
    else:
        for j in i:
            if j.lower() in vowels:
                count += 1
        if count > 2:
            complex_Word_Count += 1
    count = 0

print('complex words count =', complex_Word_Count)
```
